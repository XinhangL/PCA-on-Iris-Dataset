```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris['data']
y = iris['target']
```

```python
# Fitting PCA directly
pca = PCA()
pca.fit_transform(X)

# Accumulate all reservations until the same dimension as the original data
explained_var = np.cumsum(pca.explained_variance_ratio_)

# Plotting the amount of variation explained by PCA with different numbers of components
plt.plot(list(range(1, len(explained_var)+1)), explained_var)
plt.title('Amount of variation explained by PCA', fontsize = 14)
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
plt.show()
```

```python
print('Before pca: \n', X[:3,:])

# according to the graph, only retained 0.98 variance

pca = PCA(0.98)
X_red = pca.fit_transform(X)

# comnpare the results of the first three rows
print('After pca: \n', X_red[:3, :],'\n')

# check the percentage of each characteristic value
print('explained_variance_ratio_: ', pca.explained_variance_ratio_)

# Linear transformation rules
print('components_: ', pca.components_)
```

```python
# visualization
plt.figure(figsize=(7,7))    
for i in range(3):
    plt.scatter(x=X_red[np.where(y==i),:][0][:,0], y=X_red[np.where(y==i),:][0][:,1], alpha=0.8, label='type%s' % i)
plt.legend()
plt.show()
```
